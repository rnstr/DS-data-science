# Penentuan Hipotesis

uji hipotesis sisi kiri

H0 : harga rumah <= harga tahun sebelumnya

H1 : harga rumah > harga tahun sebelumnya

Î± = 5% = 0.05


```python
y2006 = ndf[ndf['YrSold']==2006]
y2007 = ndf[ndf['YrSold']==2007]
y2008 = ndf[ndf['YrSold']==2008]
y2009 = ndf[ndf['YrSold']==2009]
y2010 = ndf[ndf['YrSold']==2010]
```

```python
anova_test = st.f_oneway(y2006['SalePrice'],y2007['SalePrice'],y2008['SalePrice'],y2009['SalePrice'],y2010['SalePrice'])
```

```python
anova_test.pvalue
```

```python
if anova_test.pvalue>0.05:
    print('Ho diterma')
else:
    print('Ho ditolak')
```

kesimpulan : ho diterima, sehingga harga rumah tidak selalu naik tiap tahunnya.

```python
dfBankFilter = dfBank[['campaign', 'y']]
dfBankFilter.fillna(0)
dfBankFilter
```

Apakah jumlah campaign mampu meningkatkan keinginan user untuk membuka akun deposit atau tidak

a. Apa nama experimentnya? menentukan hubungan antara jumlah campaign dengan keinginan user untuk deposit    

b. Definisikan Hipotesis ?
null_hypothesis = "tidak ada hubungan antara jumlah campaign dengan keinginan user untuk deposit"
alternate_hypothesis = "ada hubungan antara jumlah campaign dengan keinginan user untuk deposit"

c. Siapa participant-nya? user bank

d. Variabel yang akan diuji apa? campaign

e. Metrics apa yang akan digunakan? jumlah campaign

f. Berapa sample size dan durasi experiment? 45211,  0.030620574951171875 detik


```python
from scipy.stats import chi2_contingency, chi2
```

```python
import time
start_time = time.time()

observed_values = pd.crosstab(dfBankFilter["campaign"], dfBankFilter["y"]).values

# calculate expected frequencies & chi square statistic
chi2_statistic, p_value, dof, expected_values = chi2_contingency(observed_values, correction = False)
print(chi2_statistic, p_value)

# find the critical value for our test
critical_value = chi2.ppf(1 - 0.05, dof)
print(critical_value)

if p_value >0.05:
    print('Ho diterma')
else:
    print('Ho ditolak')

print("--- %s seconds ---" % (time.time() - start_time))    
```

Pengujian Hipotesis
a. Penentuan hipotesis
null_hypothesis = "tidak ada hubungan antara jumlah campaign dengan keinginan user untuk deposit" alternate_hypothesis = "ada hubungan antara jumlah campaign dengan keinginan user untuk deposit"

b. Pengujian yang digunakan : Chi-Square Test

c. Hasil pengujian hipotesis : Ho ditolak

d. Kesimpulan dan interpretasi : ada hubungan antara jumlah campaign dengan keinginan user untuk deposit
